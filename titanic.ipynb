```python
import pandas as pd
```

```python
data = pd.read_csv("data/train.csv")
```

```python
data.head()
```

```python
# columns not useful :- ['PassengerId', 'Name', 'Ticket', 'Cabin']

data = data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Fare'], axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.describe()
```

```python
# lets try by dropping columns with null age 
data = data.dropna()
data.describe()
```

```python
data.shape
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() 
```

```python
data['Sex']= le.fit_transform(data['Sex']) 
```

```python
data.head() # male =1 and female = 0
```

```python
data['Embarked'] = le.fit_transform(data['Embarked'])
```

```python
data.head() # s=2, c=0, q=1
```

```python
data.describe()
```

```python
from sklearn import preprocessing
data['Age'] = preprocessing.scale(data['Age'], with_mean=True)
```

```python
data.head()
```

```python
x = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
y = data['Survived']
```

```python
x.head()
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_reg = Lasso(normalize=True)
lasso_reg.fit(x, y)
```

```python
lasso_reg.score(x, y)
```

```python
test = pd.read_csv('data/test.csv')
```

```python
test = test.dropna('index')
```

```python
test = test.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Fare'], axis=1)
test.head()
```
